#Funzione di salvataggio del render dell'environment

```python
import cv2
import numpy as np

def save_frames_to_video(frames, output_path, fps=30, frame_size=(640, 480)):
    """
    Saves a list of RGB frames to a video file.

    Args:
    - frames: List of RGB images (numpy arrays).
    - output_path: Path to the output video file.
    - fps: Frames per second of the output video.
    - frame_size: Size of each frame (width, height).
    """
    # Define the codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # You can change 'mp4v' to other codecs depending on the file extension
    out = cv2.VideoWriter(output_path, fourcc, fps, frame_size)

    # Write frames to the video
    for frame in frames:
        out.write(cv2.cvtColor(frame, cv2.COLOR_RGB2BGR))  # Convert RGB to BGR, as OpenCV expects BGR
    out.release()
```

#Inizializzazione dell'environment

```python
import gymnasium as gym
import numpy as np
env = gym.make("LunarLander-v2", render_mode="rgb_array")
observation, info = env.reset(seed=42)
```

#Esecuzione dell'ambiente per 1000 step

```python
frame_list = []
for _ in range(1000):
  action = env.action_space.sample()  # Qui prendiamo un'azione a caso, in seguito qui andrà inserita la policy
  observation, reward, terminated, truncated, info = env.step(action)
  #Step necessari per salvare il video
  frame = env.render()
  frame = np.asarray(frame, dtype=np.uint8)
  frame_list.append(frame)
  if terminated or truncated:
    observation, info = env.reset()
```

#Salvataggi

```python
save_frames_to_video(frame_list, 'lunar.mp4', fps=30, frame_size=(600, 400))
#reproduce_video('lunarlander.mp4')
env.close()
```
